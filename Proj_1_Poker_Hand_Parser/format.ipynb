```python
def parse_hand_id(hand_text: str):
    m = re.search(r"Hand #(\d+)",hand_text)
    if m:
        return m.group(1)
    return None
```

```python
parse_hand_id(hands[1])
```

```python
def parse_hole_cards(hand_text:str):
    #Expresion regular ./ChatGPT
    s = re.search(r"(?:Dealt to|Repartidas a)\s+([^\[]+)\s+\[([2-9TJQKA][shdc]\s+[2-9TJQKA][shdc])\]",hand_text)
    if s:
        return s.group(2) , s.group(1)
    return None
```
